```python
# Function to create a chromosome (5-bit binary string)
def createChromosome(n):
    return ''.join(str(random.randint(0, 1)) for i in range(n))
 
# Convert binary string to integer
def binaryToInt(binary_string):
    return int(binary_string, 2)
 
# Fitness function to calculate integer value (higher is better)
def fitness(chromosomes):
    fitness_values = {}
    for chromosome in chromosomes:
        fitness_values[chromosome] = binaryToInt(chromosome)
    return fitness_values
 
# Selection function to pick the two best chromosomes
def selection(fitness_values):
    sorted_chromosomes = sorted(fitness_values.items(), key=lambda x: x[1], reverse=True)
    return sorted_chromosomes[0][0], sorted_chromosomes[1][0]
 
# Perform crossover at 3 random points
def crossover(parent1, parent2):
    points = sorted(random.sample(range(1, len(parent1)), 3))  # 3 random crossover points
    child1, child2 = list(parent1), list(parent2)
    for i in range(len(points)):
        if i % 2 == 0:  # Swap segments
            child1[points[i]:points[i+1] if i+1 < len(points) else len(parent1)] = parent2[points[i]:points[i+1] if i+1 < len(points) else len(parent2)]
            child2[points[i]:points[i+1] if i+1 < len(points) else len(parent1)] = parent1[points[i]:points[i+1] if i+1 < len(points) else len(parent2)]
    return ''.join(child1), ''.join(child2)
 
# Main code
n = 5  # Length of binary chromosome
population_size = 3  # Number of chromosomes in the population
population = [createChromosome(n) for _ in range(population_size)]
 
# Display initial population
print("Initial Population:", population)
 
# Calculate fitness values
fitness_values = fitness(population)
print("Fitness Values:", fitness_values)
 
# Select the top two chromosomes
parent1, parent2 = selection(fitness_values)
print("Selected Parents:", parent1, parent2)
 
# Perform crossover
child1, child2 = crossover(parent1, parent2)
print("Children:", child1, child2)
```
